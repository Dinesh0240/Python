````python
n = int(input("enter n:"))
print("No. of lines:",n)
line = n

for i in range(1,n+1):
    temp = i
    for j in range(1,1+n):
        if j<line:
            print(end='- ')
    for k in range(1,i+1):
        print(temp,end=' ')
        temp-=1
    line-=1
    print()
````
