Nested Exception Handling

```python
try:
    print("first")
    x=10
    print("second line")

    try:
        print("inside nested try:")
        print(x[4])
    except:
        print("found an error inside the nested try and we handle it")

    print("third line")
except TypeError as e:
    print("Found an Type error:",e)

```

finally
-> always executed
-> after finally no code is executed

```python
try:
    print("first")
    x=10
    print("second line")

    try:
        print("inside nested try:")
        print(x[4])
    except:
        print("found an error inside the nested try and we handle it")

    print("third line")
except TypeError as e:
    print("Found an Type error:",e)
finally:
    print("finally error handled")

```

else
-> executed when there is no exception
->  Else block will execute only when no exception occurs.
    The code enters the else block only if the try clause does not raise an exception. 

```python
try:
    print("first")
    x=10
    print("second line")

    try:
        print("inside nested try:")
        print(x[4])
    except:
        print("found an error inside the nested try and we handle it")

    print("third line")
except TypeError as e:
    print("Found an Type error:",e)

else:
    print("no exception has occured")
finally:
    print("finally error handled")
```
