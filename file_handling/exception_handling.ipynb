Exception Handling
-> an unexpected or unwanted event/condition that affects the flow of the program
-> generally there can be two things -
        -> exception - which can be recoverable when it affect the flow of program
        -> error - 

-> there are two main blocks --
    try:
        ----- here the code is placed in which error can occur
    except:
        ----- here what error has occured ,it ia handled here

```python
try:
    print("first line")
    x =10
    print("second line")
    print(x[0])  #if error occured then it goes to except but if not then the code is executed
    print("third line")
except:
    print("found an error")
```

to print the type of error in the code

```python
try:
    print("first line")
    x =10
    print("second line")
    print(x[0])  
    print("third line")
except Exception as e:
    print("found an error:",e)
```

```python
try:
    print("first line")
    x =10
    print("second line")
    print(x+"8")  
    print("third line")
except Exception as e:  #Exception - here in it all the exceptions are written so we use it to find the particular error
    print("found an error:",e)
```
