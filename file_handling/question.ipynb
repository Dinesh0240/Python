# create a funcation name as palendrome check
# which take a string as a parameter 
# and return True or False wheter the string is palendrome or not 

<!-- create a exception handling 
if this s as input is a integer -->

```python
def palindrome_check(s):  
    try:
        i = 0
        j =len(s)-1

        while i<len(s):
            if s[i]!=s[j]:
                # print(s[i],s[j])
                return False
            i+=1
            j-=1
        return True

    except:
        print('found an error')

print(palindrome_check('racecar'))
# s = 'abcdef'

```
