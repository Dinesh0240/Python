File Handling

-> a mechanism to read/write or modify the files
-> open
   read/write
   save and close
-> it has two syntax
    -> open(file_name,mode)
       read/write
       .close()
    -> with open(file_name,mode) as name
            read/write

```python
f = open('user_name.txt')           #old syntax
print(f.read())
f.close()
```

-> Read mode

```python
with open('user_name.txt') as f:            #new syntax
    print("---- with new syntax----")
    print(f.read())
```

-> Write Mode

```python
with open('user_name.txt',"w") as f:            #new syntax
    print("---- with new syntax----")
    print(f.read())         #when we re writing in the file then can not read so it gives error
```

```python
with open('user_name.txt',"w") as f:            #new syntax
    print("---- with new syntax----")
    f.write("hello world")
    
```

with r+
-> previous data is not deleted 
-> both read and write

```python
with open('user_name.txt',"r+")as f:
    print(f.read())
    f.write("hello")
    
```

```python
with open('user_name.txt',"r+") as f:
    print(f.tell())
    f.write("##")
    print(f.tell())
    print(f.read())
```

with w+
-> delete the previous data 
-> both read and write

```python
with open('user_name.txt',"a+") as f:
    f.write("1")
    print(f.tell())
    # f.write("# 1")
    # print(f.tell())
    # print(f.read())
    # print(f.tell())
```

wb mode
-> open the file
->erases teh existing content
->creates file if not there
->works with binary data bytes not text

->when to use wb

    ->Use wb when working with:
        ->images 🖼️
        ->videos 🎥
        ->audio files 🎵
        ->PDFs 📄
        ->any non-text data

```python
with open('user_name.txt',"wb") as f:

    print(f.tell())
    f.write(b"Hello")
    # f.write("Hello".encode())
    print(f.tell())
```

readline()
-> use to read a line
->gives a string in output

```python
with open('user_name.txt') as f:
    print(f.tell())
    print(f.readline())
    print(f.readline())
    print(f.readline())
    print(f.readline())
    print(f.readline())
    
```

readlines()
-> gives output in the form of list of all the lines

->seek() is used to place the cursor at position/place you want
->tell() is used to give the position of the cursor where it is now

```python
with open('user_name.txt') as f:
    print(f.tell())
    print(f.readlines())
    print(f.seek(0))
    print(f.tell())
    print(f.seek(3))
    print(f.tell())
```

Loop on file to access all lines 

```python
with open("user_name.txt","r")as f:
    for i in f:
        print(i,end="")
```

```python
with open("user_name.txt","r")as f:
    for i in f:
        print(i.strip())
```
