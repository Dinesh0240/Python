Dictionary comprehension
    -->Dictionary comprehension is a concise way to create a dictionary in a single line using a loop (and  optionally a condition).

    --> syntax-
        {key: value for item in iterable}

```python
s= {n:n for n in range(0,5)}
print(s)
```

```python
d={1:'a',2:'b',3:'c'}
a = {n:n*2 for n in d }
print(a) 
```

to swap the key and values

```python
d = {"a": 1, "b": 2}
swapped = {v: k for k, v in d.items()}
print(swapped)
```

Key takeaway
d → gives keys only
d.values() → gives values only
d.items() → gives (key, value) pairs ✅ (needed here)

```python
d = {"a": 1, "b": 2}
print(d.items())
```
