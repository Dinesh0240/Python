```python
s = "SARANSH"
occurance = {}

for i in range(len(s)):
    curr = s[i]
    if curr not in occurance:
        occurance[curr] = 1
    else:
        occurance[curr] = occurance[curr]+1
    print(occurance)
```

```python
s = "data"    #one 
occurance = {}

for i in range(len(s)):
    curr = s[i]
    if curr not in occurance:
        if curr == "a" or curr == "e"or curr == "i" or curr == "o" or curr ==  "u":
            occurance[curr] = 1
    else:
        occurance[curr] = occurance[curr]+1
    print(occurance)
```

```python
s = "data"   #second
occurance = {}

for i in range(len(s)):
    curr = s[i]
    if curr not in occurance:
        if curr in "AEIOUaeiou":
            occurance[curr] = 1
    else:
        occurance[curr] = occurance[curr]+1
    print(occurance)
```
