Q2. Word Frequency Counter

-->Split is used to convert a string    into a list based on the separator given 

-->And if separator not given then it takes by default whitespaces(spaces,tabs,newline) as separator

```python
s = 'hello world hello'
l = s.split() 
d={}

for i in l:
    if i not in d:
        d[i]=1
    else:
        if i in d:
            d[i]=d[i]+1
print(d)
```
