Prime number

```python
is_prime = True
n = int(input("enter a number:"))

if n>1:
    for i in range(2,n):
        if n%i == 0:
            is_prime = False
    if is_prime :
        print("this is a prime number")
    else:
        print("Not a prime number")
else:
    print("Please enter a value greater than 1")
```

Factorial of a number

```python
num = int(input("enter num:"))
print("number:",num)
factorial = 1

for i in range(num,0,-1):
    factorial = factorial*i
print("factorial:",factorial)

```

Break statement -> it is used to stop the loop or we can say break the loop


```python
for i in range(1,11):
    if i ==6:
        break
    print(i)
```

Continue Statement -> it is used to skip the particular value/iterstion or certain condition   which we want 

```python
for i in range(1,11):
    if i ==3:
        continue
    print(i)
```

Print the numbers skipping 15 in the range of 1-21

```python
for i in range(1,21):
    if i == 15:
        continue
    print(i)
```

Print numbers upto 15 and skiping next 5 numbers in range 1-20

```python
for i in range(1,21):
    if i == 16:
        break
    print(i)
```
