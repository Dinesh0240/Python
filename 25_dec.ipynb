```python
# 2 pointer approach

s = 'abcdefg'
index = 0
last = len(s) - 1

while index < len(s):
    print("index:",index, s[index],last, s[last])
    index+= 1
    last -= 1
```

```python
# palindrome program

s = "racecar"
index = 0
last = len(s) - 1
a = 1

while(index < len(s)):
    if s[index] != s[last]:
        a = 2
        break
    index += 1
    last -= 1
if a == 1:
    print("palendrome",a)
else:
    print("not palendrome",a)
```
