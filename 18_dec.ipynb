```python
n = int(input("enter n:")) 

for i in range(n,0,-1):
    for j in range(1,n+1):
        if j<i:
            print(' ',end=' ')
    for j in range(1,n+1):
        if j==1 or j==n or i==n or i==1:
            print("*",end=' ')
        else:
            print(" ",end=' ')
    print()
```

```python
n=4
temp =1

for i in range(1,n+1):
    for j in range(n,0,-1):
        if j<=i:
            print(temp,end=' ')
            temp+=1
        else:
            print("-",end=' ')
        
    print()
```

```python
n = 4
temp =0

for i in range(1,n+1):
    temp+=1
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(temp,temp+i):
        print(j,end=' ')
    for k in range(temp+i-1,temp,-1):
        print(k-1,end=' ')
    print()
```
