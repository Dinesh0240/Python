-- here we can not acces the instance varible becoz there is no self keyword in it
-- it is like a normal function 
-- it is also called utility method

    @staticmethod

```python
class Account:
    organisation = 'ABC'
    def __init__(self,salary):
        self.salary = salary

    def info(self):
        print(self)
        print(self.salary)
        print(self.organisation)

    @staticmethod
    def info_1(amount): # yaha pr amount parameter ki value de rha h
        # amount = amount*100
        print("info 1:",amount)
        print(Account.organisation)

a1= Account(456)
a1.info()
a1.info_1(123)
```

class method
-- eska mtln ki jo class call kr rhi  h uska reference ya address jata h cls m

```python
class Amount:
    def __init__(self,salary):
        self.salary=salary

    def __repr__(self):
        return f"Amount class have salary {self.salary}" 
    
    @classmethod
    def display(cls,new):
        return cls(new)
    
print("display_1::",Amount.display(99))

class Amount_2(Amount):
    def __init__(self,salary):
        super().__init__(salary)

    def __repr__(self):
        return f"Amount_2 class have salary {self.salary}"

a2=Amount_2(101)
print("display_2::",Amount_2.display(800))
```

```python
class Amount:
    def __init__(self,salary):
        self.salary=salary

    def __repr__(self):
        return f"Amount class have salary {self.salary}" 
    
    @classmethod
    def display(cls,new):
        return cls(new)
    
print("display_1::",Amount.display(99))

class Amount_2(Amount):

    def __repr__(self):
        return f"Amount_2 class have salary {self.salary}"

print("display_2::",Amount_2.display(800))
```
