Constructor 
--> function used to define the object
--> a function used to create a object
--> to initialize the memory address for the object

--> every time when we create a function a contructor is    created

```python
class car_template:
    
    def __init__(self):     #self stores the memory address off that object
        print("constructor runn")

maruti = car_template() #when we call class then by default constructor will run
print(car_template)
print(maruti)
alto = car_template()
print(alto)
```

```python
class car_template:
    def __init__(self, gadi_color):
        print("constructor runn::==>")
        self.color = gadi_color         #instance variable

maruti = car_template("yellow")
print(maruti.color)

tarzen = car_template("purple")
print("tarzen::==>",tarzen)
print("tarzen::==>",tarzen.color)
```

```python
class student:
    #here memory is used more due to the school name is same and memory is assigned for every student 
    def __init__(self,name,id,school,fees,email):
        print("student details::==>")
        self.name = name
        self.id = id
        self.school = school
        self.fees = fees
        self.email = email

student_1 = student("rahul",786,'efg',7896,'def@gmail.com')
print(student_1.id)
print(student_1.name)
print(student_1.school)
print(student_1.email)

student_1 = student("rahul",786,'efg',7896,'def@gmail.com')
```
