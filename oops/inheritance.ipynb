```python
class elderbrother:  #from where things to be accessed is parent\base\super class
    salary = 10000

class ram(elderbrother):   #which want to access is child\derived\sub class
    pocket=400

r1=ram()
print(r1.pocket, r1.salary)
```

```python
class elderbrother:  #from where things to be accessed is parent\base\super class
    salary = 10000

class ram(elderbrother):   #which want to access is child\derived\sub class
    pocket=400
    def raminfo(self):
        print("ram function", self.pocket, self.salary)

r1 = ram()
print(r1.pocket)
```

```python
class zomato_driver:
    company='zamoato pvt ltd'

class zomato_customer(zomato_driver):
    pass

d1=zomato_driver()
print(d1.company)

d2=zomato_customer()
print(d2.company)
```

```python
class zomato_driver:
    company='zamoato pvt ltd'   #this is class varivel
    def __init__(self,name,email,ride): #self,name,email,ride are local can not be 
        self.name=name                 #accessed directly wihtout there instance
        self.email=email            #they can be accesed only by the instance created with the self
        self.totalride=ride         #keyword


class zomato_customer(zomato_driver):
    
    def __init__(self,name,email,ride):
        self.c_name=name
        self.c_email=email
        self.c_ride=ride

d1=zomato_driver('aman','aman.sharma@gmail.com',100)
print(d1.company)
print("rides are::", d1.totalride)

d2=zomato_customer('sam','sam@gmail.com',30)
print(d2.c_ride)
```

--- important to remember ---

    class variable can be accesed by both class name and object 
    but instance can only be accesed by object

```python
class zomato_driver:    
    company='zamoato pvt ltd'   #this is class varibel
    def __init__(self,name,email,ride): #self,name,email,ride are local variable can not be 
        self.name=name                 #accessed directly wihtout there instance
        self.email=email            #they can be accesed only by the instance created with the self
        self.totalride=ride         #keyword with the help of the object


class zomato_customer(zomato_driver):
    
    def __init__(self,x,y,z):
        self.c_name=x       #self.c_name is a instance varible (self --> object)
        self.c_email=y
        self.c_ride=z

d1=zomato_driver('aman','aman.sharma@gmail.com',100)
print(d1.company)
print("rides are::", d1.totalride)

c1=zomato_customer('sam','sam@gmail.com',10)
print("customer details:",c1.c_name)
print("customer details:",c1.c_email)
print("customer details:",c1.c_ride)
```
