Super keyword

--> to access/call the parent class constructor
--> it gives us the code reusability feature

```python
class zomato_driver:    
    company='zamoato pvt ltd'   
    def __init__(self,name,email,ride): 
        self.name=name                 
        self.email=email            
        self.totalride=ride         


class zomato_customer(zomato_driver):
    
    def __init__(self,x,y,z):
       print("----- inside the customer constructor ----")
       super().__init__(x,y,z)
       print("------ after calling the parent class constructor from customer ----")


# d1=zomato_driver('aman','aman.sharma@gmail.com',100)
# print(d1.company)
# print("rides are::", d1.totalride)

c1=zomato_customer('sam','sam@gmail.com',10)
print(c1.totalride)
print(c1.name)
```

```python
class driver:
    def __init__(self,name,email,trip):     #this is a example of single inheritance since there is only class
        self.name = name
        self.email=email
        self.trip=trip

    def display(self):
        print("Driver class object:",self.name,self.email)

class customer(driver):
    def __init__(self,name,email,trip):
        super().__init__(name,email,trip)

c1 = customer('jussy', 'shiv@gmail.com', 32)
print(c1.name)
print(c1.email)
print(c1.trip)

c1.display()  #here c1 is the object of customer class(child class) which can call display(parent class function)
```

```python
class driver:
    def __init__(self,name,email,trip):     #this is a example of single inheritance since there is only class
        self.name = name
        self.email=email
        self.trip=trip

    def display(self):
        print("Driver class object:",self.name,self.email)

class customer(driver):
    def __init__(self,name,email,trip):
        super().__init__(name,email,trip)

    def customer_display(self):
        print("information:",self.name,self.email,self.trip)

d1 = driver('shubham','shiv@gmail.com',189)
print(d1.name)
d1.customer_display #can not access this because d1 is the object of parent class and it is trying to access the child class which is not possible


c1 = customer('jussy', 'shiv@gmail.com', 32)
print(c1.name)
print(c1.email)
print(c1.trip)

c1.display()  #here c1 is the object of customer class(child class) which can call display(parent class function)
```

```python
class a:
    x = 'hello'

class b(a):
    y = 'hey'

class c(b):
    z = 'hii'

a1=a()
print("class a:",a1.x)
# print("class a:",a1.y)
# print("class a:",a1.z)

b1=b()
print("class b:",b1.x)
print("class b:",b1.y)
# print("class b:",b1.z) # cannot access this(z) because b is the parent class of c so parent can not access the child class

c1=c()
print("class c:",c1.x)
print("class c:",c1.y)
print("class c:",c1.z)
```

```python
class a:
    def __init__(self,x):
        self.x = x

class b(a):
    def __init__(self,y):
        self.y = y
        super().__init__(y)
class c(b):
    def __init__(self,z):
        self.z = z
        super().__init__(z)

a1=a('hey')
print("class a:",a1.x)

b1=b('hello')
print("class b:",b1.x)
print("class b:",b1.y)
# print("class b:",b1.z) # cannot access this(z) because b is the parent class of c so parent can not access the child class

c1=c('hii')
print("class c:",c1.x)
print("class c:",c1.y)
print("class c:",c1.z)
```
