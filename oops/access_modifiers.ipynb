public  access modifier
    -- can be accessed anywhere

```python
class a:
    salary = 1000

class b:
    amount=5000
    print(a.salary)

b1=b()
```

protected access modifier
    --> only sub class can access


```python
class a:
    _salary = 1000

class b:
    amount=5000
    print(a._salary)    #here without any relation we are accessing the proctected class variable

b1=b()
```

private access modifier
-- can not be accesed outside the class only inside the class it can be accessed

```python
class a:
    _salary = 1000  #protected variable
    __price = 200   #private variable

    def info(self):
        print("info::> price:",self.__price)

a1 = a()
a1.info()
```

to access the private variable outside the class also

```python
class a:
    _salary = 1000  #protected variable
    __price = 200   #private variable

    def info(self):
        print("info::> price:",self.__price)

a1 = a()
print(a1._a__price) #__price h woh variable change ho jata h if we use _class name
```
