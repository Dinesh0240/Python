method overriding

-- parent class and child class have the same methods but we override it (but different working)

```python
class parent:
    def salary(self):
        print("salary by doing job")

class child(parent):
    def salary(self):
        print("salary by doing business")# child class is dependent on itself so not need to depend on parent class for same function so it is overrided

c1=child()
c1.salary()
```

create a class as bank
    you have here a constructor name as amount
create another class as BOB bank
    you have same here a constructor varaible name as amount

bank class have a method display which will print the amount
the BOB bank aslo have display method same working as in bank class

```python
class bank:
    def __init__(self,amount):
        self.amount=amount

    def display(self):
        print("amount of bank:",self.amount)


class bob_bank(bank):
    def __init__(self,amount):
        self.amount=amount
    
    def display(self,amount):
        print("amount of bob bank:",self.amount)
        print("amount when function called:",amount)

b1=bob_bank(10)
b=bank(1)
b1.display(20)
b.display()
```

```python
class bank:
    def __init__(self,amount):
        self.amount=amount
    def display(self):
            print("bank",self.amount)
class bob_bank(bank):
    def __init__(self, amount):
        super().__init__(amount)
    def display(self):
            print("bob_bank",self.amount)

person1=bank(2000)
person2=bob_bank(1000)
person1.display()
person2.display()
```

operator overloading

```python
class vector:
    def __init__(self,i,j,k):
        self.i=i
        self.j=j
        self.k=k

    def __str__(self):
        return f"{self.i}i + {self.j}j + {self.k}k"
    
    
v1 = vector(3,4,5)
print(v1)

v2 = vector(1,2,3)
print(v2)

print(v1+v2)
```
