To update the value of a particular object

```python
class car_template: #car_template is a class
    color ="red"
    price =19000
    seat = 4

maruti = car_template()
alto = car_template()
maruti.price = 25000
print("maruti price:",maruti.price,"alto price:",alto.price)
```

when changes made in the class they are also done in its object also if they(object) does not leave the template 

```python
class car_template: #car_template is a class
    color ="red"
    price =19000
    seat = 4

maruti = car_template()
alto = car_template()
car_template.price = 30000
maruti.price = 25000
print("maruti price:",maruti.price,"alto price:",alto.price)
```

```python
class car_template: #car_template is a class
    color ="red"
    price =19000
    seat = 4

    def car_detail(self):   #function created to print car details
        print("car ==>",self.color, self.price, self.seat)

maruti = car_template()
maruti.car_detail()

alto = car_template()
alto.price=25000        #updated
alto.car_detail()
```
