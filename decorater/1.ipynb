=> Decorator
    ->design pattern/mechanism
    ->through which we can run a function and behave differently without changing the code
    ->it makes sure that function behaves differently without changing the code

```python
def outer_fun(num):
    
    def inner_fun():
        print("inner fun")

    print(inner_fun)
    return num + 10
 
a = outer_fun(15)
print("returned from fun:",a)
```

=> Inclose scope
    -> means outer variable of function can be used inside the inner function

```python
def outer_fun(num):

    def inner_fun():
        print("hey inner function this side::=>")
        y = num()
        y = y.upper()
        return y

    return inner_fun

# @outer_fun    #=>decorator which changes the behaver of the function without
def test():     #changing the code of the function
    return "my name is dinesh"

print(test())
test = outer_fun(test)
print(test())
```
