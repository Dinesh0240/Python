initialise a var 
condition -> divisible by 3 and 7
             divisible by 3 and 7 and also even
             divisible by 3 and 7 and also odd
             and at last if anything not then not divisible by any

```python
var = int(input("enter the number: "))
print(f"var: {var}")

if var % 3 == 0 and var % 7 == 0 and var % 2 ==0:
    print(f'var: {var} is divisible by 3 and 7 and is even')
elif var % 3 == 0 and var % 7 == 0 and var % 2 !=0:
    print(f'var: {var} is divisible by 3 and 7 and is odd')
elif var % 3 == 0 and var % 7 == 0:
    print(f'var: {var} is divisible by 3 and 7')
else:
    print(f'var: {var} is not divisible by 3 and 7')
```
