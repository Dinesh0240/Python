```python
# for spaces -> n -i
# star ->2*i-1
n = 5
for i in range(1,4+1):
    # printing the spaces
    for j in range(n-i):
        # printing the stars
        print(' ',end=' ')
    for j in range(2*i-1):
        # printing the row
        print('*',end=' ')
    print()
```

```python
# hollow traingle

n = int(input("enter n:"))

for i in range(1,n+1):
    for j in range(1,n+1):
        if j==1 or i==n or j==i:
            print("*",end=" ")
        else:
            print("-",end=" ")
    print()
```

```python
# right aligned hollow triangle 

n = int(input("enter n:"))

for i in range(1,n+1):
    for j in range(n,0,-1):
        if i==n or i==j or j==1:
            print("*",end=" ")
        else:
            print(' ',end=" ")
    print()
```
