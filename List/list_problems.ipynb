```python
#Check if a list is a palindrome
l = [1,2,3,2,1]    #case = 1
s= 0
e = len(l)-1
check = 0

while s<e:
    # print(s,e)
    if l[s]!=l[e]:
        check = 1
        break
    s+=1
    e-=1

if check == 0:
    print("true")
else:
    print("false")
```
