```python
prices = [7,10,5,3,6,4]

s = 0
e = len(prices)-1

while s<e: # if the loop will run till the end then the values again be the same not intercahnge so we use this condition
    print(prices)
    temp = prices[s]
    prices[s]=prices[e]
    prices[e]=temp

    print("Index:",s,e,"Reversed list values:",prices[s],prices[e],prices)
    s+=1
    e-=1
```

```python
prices = [7,1,5,3,6,4]
buy = 0
sell = 1
profit = 0

while (buy<len(prices)-1):
    print("index:",buy,sell,"values:",prices[buy],prices[sell])
    print("Profit:",prices[sell]-prices[buy])
    if prices[sell]-prices[buy]>profit:
        profit = prices[sell]-prices[buy]
        # print("profit:",profit)
    sell+=1
    if sell>len(prices)-1:
        buy+=1
        sell = buy+1
print("profit:",profit)
```

```python
prices = [7,1,5,3,6,4]
buy = 0
sell = len(prices)-1
print("Prices:",prices)

while (buy<sell):
    temp = prices[buy]
    prices[buy]=prices[sell]
    prices[sell]=temp
    # print("index:",buy,sell,"values:",prices[buy],prices[sell],prices)
    buy+=1
    sell-=1
print("New Prices:",prices)
```

```python
s = "Marge, let's \"[went].\" I await {news} telegram."   #correct verion of code
new = s.lower()
# print(new)
new_string=""

for i in new:
    if 65<=ord(i)<=90 or 97<=ord(i)<=122 or 48<=ord(i)<=57 :
        new_string = new_string+i
# print(new_string)

to_check = ""
start = 0
end = len(new_string)-1

while start<len(new_string):
    if new_string[start] == new_string[end]:
        to_check = to_check+new_string[start]
    start+=1
    end-=1

if to_check == new_string:
    print("It is palindrome")
else:
    print("Not a palindrome")
```
