Q9. Find the longest consecutive sequence in a list

```python
l = [1,2,3,10,11,12,13]
count = 1
for i in range(len(l)):
    # print(l[i])
    c=l[i]
    # print(c)
    if c and c+1  in l:
        print(c+1)
        count+=1
        print("count:",count)
```

Q17. Move all zeroes to the end while preserving order

```python
n = [0,1,0,3,12]
l = len(n)

i = 0
while i<l:
    if n[i]==0:
        # print(n)
        n.append(0)
        # print(n)
        n.remove(n[i])
    i+=1
print(n)
```
