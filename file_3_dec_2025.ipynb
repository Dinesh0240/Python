Q1 Write a program to convert a string into uppercase using the upper() method.

```python
s = input("enter the string;")
print(s.upper())

```

Q2 Write a program to convert a string into lowercase using the lower() method.

```python
s = input("enter the string;")
print(s)
st = s.lower()
print(st)
```

Q3 Write a program to convert the first character of each word into capital using the
title() method.

```python
s = input("enter the string:")
print(s)
st = s.title() #title() convert all the first character of the string into capital
print(st)
```

Q4. Write a program to capitalize the first character of a string using the
capitalize() method.

```python
s = input("enter the string:")
print(s)
st = s.capitalize()
print(st)
```

Q5. Write a program to remove the leading and trailing spaces from a string using
strip().

```python
s = input("enter the string:")
print(s)
st = s.strip()
print(st)
```

Q6. Write a program to remove spaces only from the left side using lstrip().

```python
s = input("enter the string:")
print(s)
st = s.lstrip()
print(st)
```

Q7. Write a program to remove spaces only from the right side using rstrip().

```python
s = input("enter the string:")
print(s)
st = s.rstrip()
print(st)
```
