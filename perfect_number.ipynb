```python
n = int(input("enter a num:"))
original = n
sum = 0

i = 1
while i<6:
    if n%i==0:
        sum = sum + i
    i+=1
if original == sum:
    print(f"{n} is a Perfect Number")
else:
    print(f"{n} is not a Perfect Number")
```
