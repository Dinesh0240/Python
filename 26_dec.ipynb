```python
# a string is given and we have to find the duplicate characters in it.
data = "saransh"
start = 0
end = len(data) - 1
while(start < end):
    x = data[start]
    y = data[end]
    if (data[start] == data[end]):
        print("START:",start,x,y,"END:",end)
    end-=1

    if(start==end):
        end=len(data)-1
        start+=1
        print("Reset values:","START::",start,"END:",end)
        
```
