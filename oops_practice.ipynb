super keyword
-- to reuse logic of parent class in the child class

```python
class A:
    def info(self):
        print("Info of A class")

class B(A):
    def display(self):
        super().info() # reusing the class A info method

b1=B()
b1.display()
b1.info()
```

```python
class Driver:
    def __init__(self,name,email,trip):
        self.name=name
        self.email=email
        self.trip=trip

class Customer(Driver):
    def __init__(self,name,email,trip):
        self.name=name
        self.email=email                #code rewritten
        self.trip=trip

d1=Driver('Sam',"sam@gmail.com",100)
print(d1.name,d1.email,d1.trip)

c1=Customer('jassy',"jass@gmail.com",140)
print(c1.name,c1.email,c1.trip)
```

```python
class Driver:
    def __init__(self,name,email,trip):
        self.name=name
        self.email=email
        self.trip=trip

class Customer(Driver):
    def __init__(self,name,email,trip):
        super().__init__(name,email,trip)       #reusing the code by calling the method of parent class

d1=Driver('Sam',"sam@gmail.com",100)
print(d1.name,d1.email,d1.trip)

c1=Customer('jassy',"jass@gmail.com",140)
print(c1.name,c1.email,c1.trip)
```

single level inheritance

```python
class Driver:
    def __init__(self,name,email,trip):
        self.name=name
        self.email=email
        self.trip=trip

    def display(self):
        print("driver class display",self.name,self.email,self.trip)
    
class Customer(Driver):
    def __init__(self,name,email,trip):
        super().__init__(name,email,trip)       #reusing the code by calling the method of parent class

    def customer_display(self):
        print("information:",self.name,self.email,self.trip)

d1=Driver('Sam',"sam@gmail.com",100)
print(d1.name,d1.email,d1.trip)
d1.display()
# d1.customer_display()

c1=Customer('jassy',"jass@gmail.com",140)
print(c1.name,c1.email,c1.trip)
c1.display()    #c1 can call display method of parent class and there is not display method in the customer
c1.customer_display()
```

create a class Bank
    you have a constructor having variable name as amount

create a class BOB bank
    here also same constructor as Bank and variable amount

bank class have a method display which will print the amount

bob bank also have the method and print amount


```python
class Bank:
    def __init__(self,amount):
        self.amount=amount
    
    def display(self):
        print("parent display:",self.amount)

class BOB_bank(Bank):
    def __init__(self,amount):
        self.amount=amount

    def display(self):
        print("child display:",self.amount)

b1=BOB_bank(456)
b1.display()
```
