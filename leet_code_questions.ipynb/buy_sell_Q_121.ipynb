```python
prices = [7,1,5,3,6,4]
b = 0
s = 1
profit = 0
n = len(prices)-1


while (b<n):
    # print(b,prices[b],s,prices[s])
    if prices[s]-prices[b] > profit:
        profit = prices[s]-prices[b]
    s+=1
    if s>n:
        b+=1
        s = b+1
print(profit)
```
