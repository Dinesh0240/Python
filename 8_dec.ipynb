```python
#syntax ->
for _ in range(1,6):  # only underscore(_)  special character is allowed  and name or characters 
    print(_)            # where the underscore is placed it is place holder without it the sytax is incomplete

for i in range(1,5+1):
    print(i)

for i in range(6):   # in this it is started from 0
    print(i)
```

```python
# all the odd numbers from 1-10

for i in range(1,11,2):
    print(i)

for i in range (1,11):
    if i%2!=0:
        print(i)
```

```python
# all the even numbers from 1-10
for i in range(2,11,2):
    print(i)

for i in range(1,11):  
    if i%2==0:
        print(i)
```

```python
# print numbers from 2-200 which are divible by 3 or 5 
for i in range (2,201): 
    if i%3==0 or i%5==0:
        print(i)
```

```python
# for reversing the order of the numbers:
for i in range (5 ,0,-1):
    print(i)
```

```python
# print the numbers in reverse order whicch are divisible by 7 and 3

for i in range(200,0,-1):
    if i%3==0 and i%7==0:
        print(i) 
```

```python
# sum of 1-10 numbers
sum = 0
for i in range(1,11):
    sum = sum +i
print(sum)
```

```python
# product of numbers from 5-20
prod = 1
for i in range(5,21):
    prod = prod*i
    # print(prod)
print(prod)


prod = 1
for i in range(20,4,-1):
    prod = prod*i
    # print(prod)
print(prod)
```

```python
# product of numbers from 5-25
prod = 1
for i in range(5,26):
    if i%2!=0:
        prod = prod*i
        print(prod)
print(prod)
```

```python
# sum of numbers from 50-75
sum = 0
for i in range(50,76):
    if i%2==0:
        sum = sum + i
        print(sum)
print(sum)
```

```python
# for even and odd numbers from 1-200
for i in range(2,101):
    if i%2==0:
        print("even number: ",i)
    else:
        print("odd number: ",i)
```
