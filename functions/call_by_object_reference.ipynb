for immutuable

```python
a = 299
print("a:",id(a))

c = a
print('c:',id(c))

c =300
print("c:", id(c), "a:", id(a))
```

for mutuable

```python
mylist = [10,20]
new_list = mylist
print(id(mylist), id(new_list))
new_list.append(200)
print(id(mylist), id(new_list), mylist , new_list)
```

for functions

```python
def func(x):
    print("inside x:",x , id(x))
    x = 20
    print("inside x updated:",x , id(x))
a = 20
print("before function: a",a, id(a))
func(a)
```

```python
def test(x):
    print("inside x:", x, id(x))
    x.pop()
    print("inside x updated:", x, id(x))
a = [10,20,30]
test(a)
print("a:",a, id(a))
```

Types of arguments passing

1.Required argument

```python
def test(a,b):
    print("a:",a, "b:", b)

test(10,20)
```

2. Positional argument

```python
def test(a,b):
    print("a:",a, "b:", b)

test(10,20)
test(20,10)
```

3. Keyword argument

```python
def test(a,b):
    print("a:",a, "b:", b)

test(10,20)
test(20,10)
test(a=10, b=20)#keyword argument
test(b=20, a=10 )#keyword argument
```

```python
def sch(u_name,college,school):
    print(f"user name:{u_name} college:{college} school:{school} ")

sch(u_name = "sam", college = "other", school = "raj school")
```

4.Variable length argument

```python
def sch(*var):      #var is variable length argument
    print(var)  #multiple arguments can be given at the same time when the function is call
    print(type(var)) #var stores data in the form of tuple

sch( "sam", "other", "raj school")  
```

even number

```python
def test(*var):
    for i in range(0,len(var)):
        if var[i]%2==0:
            print(f"{var[i]} is a even number")
test(2,3,4,5,8,161,16,13,531,153)
```
