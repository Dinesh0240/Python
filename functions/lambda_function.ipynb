Lambda function

->Major diffence between lambda and def is that in def we  have to explicity return value of function whereas
     in lambda it implicitly return the value 

->Lambda are generally used in higher order function

```python
s = lambda x: x%5
print(s(18))
```

```python
lambda x: "Even" if x % 2 == 0 else "Odd"
```

```python
check = lambda x: "Even" if x % 2 == 0 else "Odd"
print(check(4))  
print(check(7))
```

```python

def fun(x):
    print(x+5)

out = fun
# out = fun(8)
print("out:",out)

out2= lambda x: x+5
print(out2(3))
print(out)
```

```python
d = {1:lambda x:x+5}
print(d)  #returns dictionary

print(d[1]) #returns address of lambda function

print(d[1](1))  #given x a value by accessing the key of the dictionary

a = d[1]   #assign  dictionary into a variable
a(2)    #passing a argument through the variable assigned for the dictionary
```

```python
dict ={1:lambda x:x+5}
print(dict)
print(dict[1])
```
