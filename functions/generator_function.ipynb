```python
def fun():    # it is a function 
    print("line 1")

gen = fun()
print(gen)
```

```python
def fun():      # it is a generator
    print("line 1")
    yield 1

gen = fun()
print(gen)
```

```python
def fun():
    print("line 1")
    yield 1
    print("line 2")
    yield 2
gen = fun()
print(gen)
gen_1 = next(gen)
gen_2 = next(gen)
```

```python
def fun():
    print("first time")
    i =1
    while(True):
        print("Inside loop:",i)
        yield i
        print("<- Restart Generator ->Increase value for i")
        i+=1

generator_fun = fun()
print(next(generator_fun))

for i in range(3):
    print("hey")

print(next(generator_fun))
print(next(generator_fun))
```

Fibonacci series

```python
def fun():
    a=0
    b=1

    while True:
        temp=a+b
        yield f"fibonaci Series:{a}"
        b=a
        a=temp
        # print(a,b,temp)

gen=fun()
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def fun(n):         #this is one case where generator gives error because the iteration is over and there is no value where it is iterable
    a=0
    b=1

    while n:
        temp=a+b
        yield f"fibonaci Series:{a} n:{n}"
        b=a
        a=temp
        # print(a,b,temp)
        n-=1

gen=fun(5)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def fun(n):  
    a=0
    b=1

    while (n):
        temp=a+b
        yield f"fibonaci Series:{a}"
        b=a
        a=temp
        # print(a,b,temp)
        n-=1

gen=fun(7)

for i in gen:       #for loop by-default runs next function in the background
    print(i)
```
