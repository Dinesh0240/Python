filter function


```python
def fun(n):     #one way 
    return n%2==0
    
l = [1,2,3,4,5]
a = filter(fun,l)
# print(list(a))
print(a)
print(next(a))
print(next(a))
# print(next(a))
```

```python
def fun(n):         #second way
    return n%2==0
    
l = [1,2,3,4,5]
a = filter(fun,l)
print(list(a))
# print(a)
# print(next(a))
# print(next(a))
# print(next(a))
```
