first class function
-> a design pattern to reuse a function 
-> it has 4 properties:

1st property
    -> assigning to a variable

```python
def even():
    print("this is even number")

y = even    #function assigned to a variable
print("through variable:",y)    #function not called
print("function",even)
```

```python
def even():
    print("this is even number")

y = even()          #function is called
print("through variable:",y)        # variable in which function assigned 
print("function:",even())           #fuction is called
```

2th property
    -> function can be assigned to list, tuple, set, dictionary

```python
def test():
    print("this is a even number")

l = [20,30,test]
print(l)
```

```python
def test():
    print("this is a even number")

l = [20,30,test()]  #function is called 
print(l)        #list is printed with the value of function that it return (as function return nothing so none is returned)
```

```python
def test():
    print("this is a even number")

l = [20,30,test]    #here function is not called so its memory address is displayed 
print(l)
print(l[-1]())  #here function is called through the index of list so it returns none
# l[-1]()
```

```python
def test():
    print("this is a even number")

l = [20,30,test()]  #function is called so the value return is none
print(l)
print(l[-1]())  #so here function is callable as none and it is not collable
# l[-1]()
```

3nd property
    -> taking function as a argument in another function

```python
def square(n):
    print("value:",n+5)

def test(a,b):
    pass

test(a=10,b=square(5))
```

```python
def square(n):
    n = int(input("enter a num:"))
    print("value:",n+5)

def test(a,b):
    pass

test(a=10,b=square(5))
```

```python
def even():
    print("this is even number")

y = even()          #function is called
print("through variable:",y)        # variable im which fun assigned 
print("function:",even())           #fuction is called
# print(even())
```

```python
def square(n):
    print(n+5)

def test(a,b):      #a is 10  b=> memory address of square function
    print("Inside test:",a,b)
    b(23)           # calling the square  method square(23)
                    # b(23)   square(23)
test(a=10,b=square)
```

```python
def square(n):
    print(n+5)

def test(a,b):      #a is 10  b=> memory address of square function
    print("Inside test:",a,b)
    b(a)           # calling the square  method square(23)
                    # b(23)   square(23)
test(a=10,b=square)
```

```python
def square(n):
    return n*n

def test(a,b):      #a is 10  b=> memory address of square function
    print("Inside test:",a,b)
    print(a*square(a))           # calling the square  method square(23)
    # print(a*b(a))                # b(23)   square(23)
test(5,square)
```
