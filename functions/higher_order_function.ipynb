Higher order function

-> all HOF function are FCF but not all FCF are HOF
-> HOF is a subset of FCF

```python
def test(name):
    return "hello" + name

def case(fun,name):
    return fun(name) + "!"

print(case(test," sam"))
```

```python
def test(name):
    return "hello" + name

def case(name):
    return name + "!"

print(case(test(" sam")))
```

```python
def apply(func, x): # higher order function
    return func(x)

def square(n): # function to be passed
    return n * n

print(apply(square, 5))
```
