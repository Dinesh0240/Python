to change the global variable use global keyword

```python
def test():
    global num
    num = num+1
    print("inside fun;",num)

num = 19
test()
print("num:",num)
```

```python
def test(num):
    num = num+1
    print(num)

num = 19
output = test(num)
print("output value:",output)
```

```python
def prime():
    n = 5
    for i in range(2,n):
        if (n%i==0):
            return 0
    return 1

z = prime()
# print(z)
if (z==1):
    print("prime")
else:
    print("not a prime")
```
