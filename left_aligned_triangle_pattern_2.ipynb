````python
n = int(input("enter n:"))
print("No. of lines:",n)
line = n # here line is used so that the value of no. of lines does not change 

for i in range(1, n+1):
    for j in range(1,n+1):
        temp = j
        if j<=line:
            print(temp,end=' ')
        else:
            print(end='')
    line-=1
    print()
````
