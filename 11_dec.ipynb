count the number of digits

```python
n = int(input("enter number:"))
print("Number:",n)
reverse = 0

while n!=0:
    digit = n%10
    print(f'The last digit : {digit}')
    print(f'before adding last digit to the reverse var:{reverse}')
    reverse = reverse*10 + digit
    print(f'number before eliminating the last digit:{n}')
    n = n//10
print(reverse)
```

Count the number of digits in a given number

```python
n = int('456')
count = 0

while n!=0:
    count +=1
    n = n // 10
print("No. of digits in n:",count)
```

calculate factorial of a given number using while loop n = 5


```python
m = 5
factorial = 1
i = 1

while i <= m:
    factorial *= i
    i += 1

print(f"Factorial of {m} = {factorial}")
```
