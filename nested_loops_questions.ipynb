````python
# pyramid pattern
r = 4
c = 4

for i in range(1,r+1):
    for j in range(c,0,-1):
        if j>i:
            print("-",end='')
        else:
            print('*',end='')
    # print()
    for a in range(1,r+1):
        for b in range(1,c+1):
            if b<=a:
                print("*",end='')
            else:
                print('-',end='')
    print()
````

````python
# left_side_aligned_triangle_pattern

n=int(input("enter n:"))

for i in range(1,n):
    for j in range(1,1+n):
        if j<=i:
            print('*',end=' ')
        else:
            print(end='- ')
    print()
for i in range(n,0,-1):
    for j in range(1,n+1):
        if j>i:
            print('-',end=' ')
        else:
            print('*',end=' ')
    print()
````

````python
# right_side_aligned_triangle_pattern

n = int(input("enter n:"))

for i in range(1,n):
    for j in range(n,0,-1):
        if j>i:
            print("-",end=' ')
        else:
            print("*",end=' ')
    print()

for i in range(n,0,-1):
    for j in range(n,0,-1):
        if j>i:
            print("- ",end='')
        else:
            print('* ',end='')
    print()
````
