right shift

```python
print(10>>1)   #formula num//2 ** bits to shift
```

```python
print(43>>2)  #43//2**2 = 10
```

left shift

```python
print(10<<2)  #formula num*2**bits to shift
```

```python
print(3<<2) #3*2**2 =12
```

& operator

```python
print(4&12) #if both the corresponding bits are 1 only then 1 otherwise 0
```

```python
print(3&9)
```

Or(|) operator

```python
print(3|12) #if only one of the corresponding digit is 1 then output 1 else 0
```

```python
print(18|27)
```

X-OR(^) operator

```python
print(5^3) # if and if only both are same digits then 0 otherwise 1
```

```python
print(4^7)
```

```python
print(5^6)
```
