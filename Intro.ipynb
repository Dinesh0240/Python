```python
# check wheter the password is strong
# len ->8
#check wheter it has alpha
#check whether it has number
#strong password , weak password

password = 'Abcd@1243'

len_password = len(password)>=8
print(len_password)


has_digit= False
has_upper= False
has_lower= False
has_special=False

for c in password:
    if c.isdigit():
        has_digit=True
    elif c.upper():
        has_upper=True
    elif c.lower():
        has_lower =True
    elif not c.isalnum():
        has_special=True
    
if has_digit and has_lower and has_special and has_upper and len_password:
    print("strong password")
else:
    print("weak password")
```
