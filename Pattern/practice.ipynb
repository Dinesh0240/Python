```python
s = 'HeLLo'
i = 0
while i<len(s):
    if s[i].isupper():
        l = s[i].lower()
        print(l,end="")
    if s[i].islower():
        u = s[i].upper()
        print(u,end="")
    i+=1
```

```python
n = 4
for i in range(n):
    for j in range(2*n-1):
        if j<i:
            print("-",end=" ")
        else:
             if  j<n+i:
                print("*",end=" ")
    print()
```
