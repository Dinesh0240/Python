```python
s = "racecar"
exit = 0
i = 0
j = len(s)-1
while (i<len(s)):
    # print(f"index:{i} character_index:{s[i]} last_index:{j} last_character:{s[j]}")
    if s[i]!=s[j]:
        exit = 1
        break
    i+=1
    j-=1
if exit == 1:
    print("Not a palindrome")
else:
    print("palindrome")
```
