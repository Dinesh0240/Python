```python
# 450 in backward print those value upto 5 digits which are divided by 7 and 9
start = 450
count = 1

while(count<=5):
    if start%7==0 and start%9==0:
        print("Divided by 7 and 9:",start,count)
        count+=1
    start-=1
```
