2. Palindrome Words
Print all palindrome words found in a given sentence.
○ Example: "mom and dad went to civic center" -> output: mom dad civic

```python
s = "mom and dad went to civic center"
words = s.split()
print(words)

for w in words:
    rev = w[::-1]
    if rev == w:
        print(w)
```
