````python
n=int(input("enter n:"))
print("No. of lines:",n)

for i in range(1,n+1):
    for j in range(1,n+1):
        if j==1 or i==j or i==n :
            print("* ", end='')
        else:
            print(end=' ')
    print()
````

````python
n=int(input("enter n:"))
print("No. of lines:",n)

for i in range(1,n+1):
    for j in range(n,0,-1):
        if j==1 or i==j or i==n :
            print("*-", end='')
        else:
            print(end='-')
    print()
````
